```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], test_size=0.2, random_state=0)
```

```python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
```

```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
# create a scatter matrix from the dataframe, color by y_train
#sns.pairplot(df, y_train)
sns.pairplot(df, kind='scatter', diag_kind='auto', markers=None, height=2.5, aspect=1, corner=False, dropna=False, plot_kws=None, diag_kws=None, grid_kws=None, size=None)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

Imagine we found an iris in the wild with a sepal length of
5 cm, a sepal width of 2.9 cm, a petal length of 1 cm, and a petal width of 0.2 cm.
What species of iris would this be?

```python
X_new = np.array([[5.0, 2.9, 1.0, 0.2]]) #2D array as sklearn only expects 2D array
```

```python
prediction = knn.predict(X_new) #Target in which X_new belongs
prediction
```

```python
y_pred = knn.predict(X_test) #Evaluating the model using the test Dataset
print("Test set predictions:\n {}".format(y_pred))
```

```python
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```
